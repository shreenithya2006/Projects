```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv(r"C:\Users\NSTIW\Downloads\archive (13)\dataset.csv")

print(data.columns)

sns.pairplot(data, vars=['COUGHING', 'WHEEZING', 'ALCOHOL_CONSUMING', 'ANXIETY'], hue='CHEST_PAIN', palette='coolwarm')

plt.suptitle('Pair Plots of Selected Features')  
plt.show()
```

```python
plt.figure(figsize=(7, 6))
sns.violinplot(data=data, x='YELLOW_FINGERS', y='LUNG_CANCER', palette='coolwarm')
plt.title('Violin Plot: ALLERGY by YELLOW_FINGERS')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(data['PEER_PRESSURE'], data['ALLERGY'], s=data['CHEST_PAIN'] / 10, alpha=0.5, c=data['COUGHING'].apply(lambda x: 1 if x == 'M' else 0), cmap='coolwarm')
plt.title('Bubble Chart: CHEST_PAIN vs LUNG_CANCER , Sized by Area')
plt.xlabel('ALCOHOL_CONSUMING')
plt.ylabel('YELLOW_FINGERS')
plt.colorbar(label='ALLERGY (0: Benign, 1: Malignant)')
plt.show()
```

```python
from pandas.plotting import parallel_coordinates

subset = data[['ALLERGY', 'PEER_PRESSURE', 'CHEST_PAIN', 'ANXIETY', 'ALCOHOL_CONSUMING']]
plt.figure(figsize=(9, 8))
parallel_coordinates(subset, class_column='ALLERGY', color=['orange', 'blue'])
plt.title('Parallel Coordinates Plot')
plt.show()
```

```python
plt.figure(figsize=(9, 5))
sns.kdeplot(data=data, x='SHORTNESS_OF_BREATH', hue='ANXIETY', fill=True, palette='coolwarm')
plt.title('Density Plot: COUGHING by ANXIETY')
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(11, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(data['FATIGUE'], data['PEER_PRESSURE'], data['COUGHING'], c=data['ANXIETY'].apply(lambda x: 1 if x == 'M' else 0), cmap='coolwarm')
ax.set_xlabel('YELLOW_FINGERS')
ax.set_ylabel('LUNG_CANCER')
ax.set_zlabel('CHEST_PAIN')
plt.title('3D Scatter Plot')
plt.show()
```

```python
g = sns.FacetGrid(data, col="ANXIETY", height=6, aspect=1)
g.map(sns.histplot, "LUNG_CANCER", kde=True, color="blue")
g.set_titles("{col_name} ANXIETY")
plt.show()
```
